```r
library(reshape2)
library(ggplot2)
```

## Sample mean of Gaussian Distribution

```r
n = 1000
mean = 1
sd = 1
set.seed(12345)
samples = rnorm(n, mean, sd)
samples_mean = cumsum(samples)/1:n  # compute the cumulative sum
data1 = data.frame(n= 1:n, samples = samples)
data2 = data.frame(n= 1:n, samples_mean = samples_mean)

# scatter plot of samples
ggplot(data1, aes(x =n, y= samples)) + geom_point()
```

```r
ggplot(data2, aes(x =n, y= samples_mean)) + geom_point()
```

## Sample mean of Chi Square Distribution

```r
#### set.seed(888)
# Method 1: Simulate Chi square(n) from the sum of square of n iid starndard normal variables
n = 10000
mean= 0
sd = 1
samples = rnorm(n, mean, sd)
samples = samples^2
samples_mean = cumsum(samples)/1:n
data3 = data.frame(n= 1:n, samples_mean= samples_mean)

# Method2: simulate from chi square n
sample = rchisq(n, df= 1)
sample_mean = cumsum(sample)/1:n
data4 = data.frame(n= 1:n, samples_mean= sample_mean)

data = merge(data3, data4, by="n")
colnames(data) = c('id', "Indirect Simulation", "Direct Simulation")
#head(data)

data = melt(data, id= 'id')
#head(data)
# scatter plot of samples
ggplot(data, aes(x =id, y= value, color= variable)) + geom_point()
```
